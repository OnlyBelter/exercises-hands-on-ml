```python
from __future__ import absolute_import, division, print_function, unicode_literals

# TensorFlow and tf.keras
import tensorflow as tf
from tensorflow import keras
from sklearn.decomposition import PCA
from sklearn.manifold import LocallyLinearEmbedding, TSNE, MDS
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
import time

print(tf.__version__)  # 2.1.0
```

```python
mnist = keras.datasets.mnist
(train_nums, train_labels), (test_nums, test_labels) = mnist.load_data()
train_nums.shape, test_nums.shape
```

```python
some_digit = train_nums[0]
some_digit_image = some_digit.reshape(28, 28)
plt.imshow(some_digit_image, cmap='binary')
plt.axis('off')
plt.show()
```

```python
train_nums_flattened = np.array([x.flatten() for x in train_nums])
train_nums_flattened.shape
```

### reduce dimension by PCA

```python
pca = PCA(n_components=2)
X_reduced_pca = pca.fit_transform(train_nums_flattened)
X_reduced_pca.shape
```

```python
plt.figure(figsize=(15, 12))
plt.scatter(X_reduced_pca[:, 0], X_reduced_pca[:, 1], c=train_labels, label=train_labels)
plt.colorbar()
plt.title('PCA visualization of MNIST')
plt.tight_layout()
plt.savefig('mnist_pca_vis.png', dpi=300)
```

### reduce dimension by t-SNE
- about 30 min using 4 cores

```python
# tsne = TSNE(n_components=2, n_jobs=4)
# X_reduced_tsne = tsne.fit_transform(train_nums_flattened)
# X_reduced_tsne.shape
# np.save('X_reduced_tsne', X_reduced_tsne)
X_reduced_tsne = np.load('X_reduced_tsne.npy')
X_reduced_tsne.shape
```

```python
def plot_reduced_mnist(X, labels, title, file_name=None):
    plt.figure(figsize=(15, 12))
    plt.scatter(X[:, 0], X[:, 1], c=labels)
    plt.colorbar()
    plt.title(title)
    plt.tight_layout()
    if file_name:
        plt.savefig(file_name, dpi=300)
```

```python
plot_reduced_mnist(X=X_reduced_tsne, labels=train_labels, file_name='mnist_tsne_vis.png', title='MNIST vis by t-SNE')
```

```python
# reduce dimension first by PCA, then use t-SNE
pca_95 = PCA(n_components=0.95)
X_reduced_pca_95 = pca_95.fit_transform(train_nums_flattened)
X_reduced_pca_95.shape
```

```python
t0 = time.time()
# tsne = TSNE(n_components=2, n_jobs=4)
X_reduced_tsne2 = tsne.fit_transform(X_reduced_pca_95)
print(X_reduced_tsne2.shape)
np.save('X_reduced_tsne_pca_first', X_reduced_tsne2)
t1 = time.time()
print("t-SNE took {:.1f}s.".format(t1 - t0))
```

```python
plot_reduced_mnist(X=X_reduced_tsne2, labels=train_labels, file_name='mnist_tsne_vis_reduce_by_pca_first.png', title='MNIST vis by t-SNE after reducing by PCA')
```

### reduce dimension by LLE 

```python
t0 = time.time()
lle = LocallyLinearEmbedding(n_components=2, n_jobs=4)
X_reduced_lle = lle.fit_transform(X_reduced_pca_95)
print(X_reduced_lle.shape)
np.save('X_reduced_lle', X_reduced_lle)
t1 = time.time()
print("LLE took {:.1f}s.".format(t1 - t0))
```

```python
plot_reduced_mnist(X=X_reduced_lle, labels=train_labels, file_name='mnist_lle_vis_reduce_by_pca_first.png', title='MNIST vis by LLE after reducing by PCA')
```

### reduce dimension by LDA

```python
t0 = time.time()
lda = LinearDiscriminantAnalysis(n_components=2)
X_reduced_lda = lda.fit_transform(X_reduced_pca_95, train_labels)
print(X_reduced_lda.shape)
np.save('X_reduced_lda', X_reduced_lda)
t1 = time.time()
print("LDA took {:.1f}s.".format(t1 - t0))
```

```python
plot_reduced_mnist(X=X_reduced_lda, labels=train_labels, file_name='mnist_lda_vis_reduce_by_pca_first.png', title='MNIST vis by LDA after reducing by PCA')
```

```python
t0 = time.time()
lda2 = LinearDiscriminantAnalysis(n_components=2, solver='eigen', shrinkage='auto')
X_reduced_lda2 = lda2.fit_transform(X_reduced_pca_95, train_labels)
print(X_reduced_lda2.shape)
np.save('X_reduced_lda', X_reduced_lda2)
t1 = time.time()
print("LDA2 took {:.1f}s.".format(t1 - t0))
```

```python
plot_reduced_mnist(X=X_reduced_lda2, labels=train_labels, file_name='mnist_lda2_vis_reduce_by_pca_first.png', title='MNIST vis by LDA2 after reducing by PCA')
```
